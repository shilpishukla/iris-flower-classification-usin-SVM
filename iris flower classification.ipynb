# iris flower classification using SVM

# Loading Dataset

## The data related to these flowers is avialable in scikit-learn.org and can be obtained by calling the function load_iris().To load the dataset,we can call load_iris() function that gives a 'Bunch' object as:

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

# Convert iris dataset into dataframe

```python
import pandas as pd
df = pd.DataFrame(iris.data, columns= iris.feature_names)
df.head()
```

# Add target column also to df

```python
df['target'] = iris.target
df.head()
```

# To display pair plots

## pair plots   hue='target'  -> change color based on target flower, this uses blue, red and green palette="brg"

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.pairplot(df, hue='target', palette="brg")
plt.show()
```

# From the above graphs, we can understand that we can separate the datapoints using a line.



# Divide the data frame into independent variables(x) and dependent variable(y) as:

## x represents all columns except the target column
## y represents the target column

```python
x = df.drop(['target'], axis = 'columns')
y = df.target
```

# Split the Input data (x) into train and test data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

# Create SVM Model

## by creating an object to SVC(Support Vector Classifier) class as:

```python
from sklearn.svm import SVC
model = SVC() # C=1.0, kernel='rbf', gamma='scale'
```

# train the SVM model

```python
model.fit(x_train, y_train)
```

# Find accuracy level

```python
model.score(x_test, y_test)
```

# take a random sample for prediction

## in the data,0 to 49 -> setosa, 50 to 99 -> versicolor, 100 to 149 -> virginica

```python
model.predict([iris.data[50]])  # array([1]) -> versicolor
```

```python
model.predict([[7.7, 2.6, 6.9, 2.3]])   # array([2]) -> virginica
```
